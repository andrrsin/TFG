```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pywib as pwb
from scipy import stats
```

```python
tf = pd.read_csv('raw/Coordenadas.tsv', sep='\t')

tf_raw=tf[['Sensor','Recording timestamp','Participant name','Gaze point X','Gaze point Y','Presented Stimulus name']].copy()
tf_filtered = tf_raw.rename(columns={'Sensor':'eventType','Recording timestamp':'timeStamp' ,'Participant name': 'sessionId', 'Gaze point X': 'x', 'Gaze point Y': 'y'})


```

```python

tf_clean = tf_filtered.dropna(subset=['x', 'y', 'Presented Stimulus name', 'sessionId'])

traces = {
    session_id: [
        trace_df.reset_index(drop=True)
        for _, trace_df in session_df.groupby('Presented Stimulus name')
    ]
    for session_id, session_df in tf_filtered.groupby('sessionId')
}



```

```python
jerkiness = pwb.jerkiness(tf_clean, traces = traces, per_traces=True)
```

```python
metrics = pwb.jerkiness_metrics(tf_clean, traces=traces)
for session, session in metrics.items():
    print(f"Session: {session}, Mean Jerkiness: {session['mean']}, Max Jerkiness: {session['max']}, Min Jerkiness: {session['min']}")

```
